```python
# Importamos la librería Pandas
import pandas as pd
# Importamos numpy
import numpy as np
# Importamos matplotlib
import matplotlib.pyplot as plt
```

```python
# Importamos el dataset del dengue
df = pd.read_csv("dengue_abierto19.csv")
```

### Analizando por Columnas
- Por la columna edad: EDAD_ANOS

```python
# Medidas de tendencia central de la columna edad "EDAD_ANOS"
print("-- Medidas de Tendencia Central de la Columna Edad --")
print("Media: {:.2f}".format(df["EDAD_ANOS"].mean()))
print("Mediana: {:.2f}".format(df["EDAD_ANOS"].median()))
print("Moda: {}".format(df["EDAD_ANOS"].mode()))
print("Desviación estándar : {:.2f}".format(df["EDAD_ANOS"].std()))



# CATEGORIZAMOS DATOS DE LA COLUMNA EDAD_ANOS EN CUATRO CATEGORÍAS
# Niños de < 12 años
# Adolescentes   >=12 y < 20
# Adultos >= 20 y <60
# Adultos mayores >= 60

niños = df[df["EDAD_ANOS"]<12].count()
#niños = len(niños) No me funcionó 🤔
niños = niños["EDAD_ANOS"]
print("Número de niños: {}".format(niños))

adolescentes = df[(df["EDAD_ANOS"]>=12) & (df["EDAD_ANOS"] < 20)].count()
adolescentes = adolescentes["EDAD_ANOS"]
print("Número de adolescentes: {}".format(adolescentes))

adultos = df[(df["EDAD_ANOS"]>=20) & (df["EDAD_ANOS"] < 60)].count()
adultos = adultos["EDAD_ANOS"]
print("Número de Adultos: {}".format(adultos))

adultosm = df[(df["EDAD_ANOS"]>=60)].count()
adultosm = adultosm["EDAD_ANOS"]
print("Número de Adultos Mayores: {}".format(adultosm))

print("-- VERIFICANDO TOTALES --")
print("* Totales por  Categorías: ", niños+adolescentes+adultos+adultosm)
print("* Total de Pacientes del Dataset:", df["EDAD_ANOS"].count())
```

```python
# Graficamos por categorías de la columna EDAD_ANOS
# Creamos dos objetos fig y ax
fig, ax = plt.subplots()
# Definimos colores para cada barra
colores = ['blue', 'orange', 'green', 'red']
# Creamos barras individuales
ax.bar('Niños', niños, 0.5, label='Niños', color=colores[0])
ax.bar('Adolescentes', adolescentes, 0.5, label='Adolescentes', color=colores[1])
ax.bar('Adultos', adultos, 0.5, label='Adultos', color=colores[2])
ax.bar('Adultos mayores', adultosm, 0.5, label='Adultos mayores', color=colores[3])
ax.set_ylabel('Cantidad')
ax.set_title('Distribución de Pacientes de Dengue por Edad')
# Agregar las cantidades encima de cada barra
for i, v in enumerate([niños, adolescentes, adultos, adultosm]):
    ax.annotate(str(v), xy=(i, v), xytext=(0, 3), textcoords="offset points", ha='center')
# Colocar la leyenda en la esquina superior izquierda
ax.legend(loc='upper left')
# Mostrar gráfica de barras
plt.show()
```

```python
# Agregamos una gráfica de barras del campo EDAD_ANOS del dataset
# Categorizar los casos por grupos de edad
# Niños menores de 12 años
# Adolescentes de 12 y menores de 20 años
# Adultos =>20 y <60
# Adulto mayor >= 60
niños = df[df["EDAD_ANOS"] < 12]["HEMORRAGICOS"].value_counts()
adolescentes = df[(df["EDAD_ANOS"] >= 12) & (df["EDAD_ANOS"] < 20)]["HEMORRAGICOS"].value_counts()
adultos = df[(df["EDAD_ANOS"] >= 20) & (df["EDAD_ANOS"] < 60)]["HEMORRAGICOS"].value_counts()
adultos_mayores = df[df["EDAD_ANOS"] >= 60]["HEMORRAGICOS"].value_counts()

# Crear la figura
plt.figure(figsize=(10, 6))

# Graficar los datos
niños = plt.bar([0, 1], niños, color='blue', alpha=0.7, width=0.2, label='Niños')
adolescentes_bars = plt.bar([0.2, 1.2], adolescentes, color='orange', alpha=0.7, width=0.2, label='Adolescentes')
adultos_bars = plt.bar([0.4, 1.4], adultos, color='green', alpha=0.7, width=0.2, label='Adultos')
adultos_mayores_bars = plt.bar([0.6, 1.6], adultos_mayores, color='red', alpha=0.7, width=0.2, label='Adultos Mayores')

# Configurar la gráfica
plt.title('Distribución de Casos de Dengue Hemorrágico y Clásico por Edad')
plt.xlabel('Tipo de Dengue')
plt.ylabel('Número de casos')
plt.xticks([0.3, 1.3], ['Dengue Clásico', 'Dengue Hemorrágico'])
plt.legend()
plt.tight_layout()

# Agregar etiquetas de número a cada barra
for bars in [niños, adolescentes_bars, adultos_bars, adultos_mayores_bars]:
    for bar in bars:
        height = bar.get_height()
        plt.annotate('{}'.format(height),
                     xy=(bar.get_x() + bar.get_width() / 2, height),
                     xytext=(0, 3),
                     textcoords="offset points",
                     ha='center', va='bottom')

# Mostrar la gráfica
plt.show()
```

```python
# Contar la cantidad de casos de dengue asignados a cada entidad
casos_por_entidad = df['ENTIDAD_ASIG'].value_counts().sort_index()

# Crear la gráfica de barras
plt.figure(figsize=(12, 6))
casos_por_entidad.plot(kind='bar', color='green')

# Configurar la gráfica
plt.title('Comportamiento del Dengue  por Entidad'+"\n" +"Febrero 19 de 2024")
plt.xlabel('Entidad de la República Mexicana')
plt.ylabel('Número de casos')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mayor legibilidad
plt.grid(axis='y')  # Mostrar líneas de la cuadrícula solo en el eje y

# Agregar etiquetas de números a cada barra
for index, value in enumerate(casos_por_entidad):
    plt.text(index, value, str(value), ha='center', va='bottom')

# Mostrar la gráfica
# Ajustamos la gráfica
plt.tight_layout()
plt.show()
```

```python
# Número de hombres y mujeres en el dataset
# Contar los valores de la columna "Sexo"
numporSexo = df["SEXO"].value_counts()
print(numporSexo)
# Crear la gráfica de pastel
plt.figure(figsize=(3, 3))
plt.pie(numporSexo, labels=['Hombres', 'Mujeres'], autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], startangle=90)
plt.title('Distribución Pacientes de Dengue Clasificado por Hombres y Mujeres')
# Para que el gráfico de pastel sea circular
plt.axis('equal')  
plt.show()
```

```python
# Contar el número de casos por categoría de HEMORRAGICOS
# Suponiendo que 1 representa la presencia de la condición
hemorragicos = df[df["HEMORRAGICOS"] == 1]
num_hemorragicos = len(hemorragicos)
print("Número de casos con hemorragias:", num_hemorragicos)

# Repite el mismo proceso para las otras variables
diabetes = df[df["DIABETES"] == 1]
num_diabetes = len(diabetes)
print("Número de casos con diabetes:", num_diabetes)

hipertension = df[df["HIPERTENSION"] == 1]
num_hipertension = len(hipertension)
print("Número de casos con hipertensión:", num_hipertension)

ulc_peptica = df[df["ENFERMEDAD_ULC_PEPTICA"] == 1]
num_ulc_peptica = len(ulc_peptica)
print("Número de casos con enfermedad ulceropeptica:", num_ulc_peptica)

enfermedad_renal = df[df["ENFERMEDAD_RENAL"] == 1]
num_enfermedad_renal = len(enfermedad_renal)
print("Número de casos con enfermedad renal:", num_enfermedad_renal)

inmunosupr = df[df["INMUNOSUPR"] == 1]
num_inmunosupr = len(inmunosupr)
print("Número de casos con inmunosupresión:", num_inmunosupr)

cirrosis_hepatica = df[df["CIRROSIS_HEPATICA"] == 1]
num_cirrosis_hepatica = len(cirrosis_hepatica)
print("Número de casos con cirrosis hepática:", num_cirrosis_hepatica)

embarazo = df[df["EMBARAZO"] == 1]
num_embarazo = len(embarazo)
print("Número de casos de pacientes embarazadas:", num_embarazo)

#grafica
fig, ax = plt.subplots()
ax.bar('Hemorragias', num_hemorragicos, 0.5, label='Hemorragias')
ax.bar('Diabetes', num_diabetes, 0.5, label='Diabetes')
ax.bar('Hipertensión', num_hipertension, 0.5, label='Hipertensión')
ax.bar('Ulceropeptica', num_ulc_peptica, 0.5, label='Enfermedad Ulceropeptica')
ax.bar('Enfermedad Renal', num_enfermedad_renal, 0.5, label='Enfermedad Renal')
ax.bar('Inmunosupresión', num_inmunosupr, 0.5, label='Inmunosupresión')
ax.bar('Cirrosis Hepática', num_cirrosis_hepatica, 0.5, label='Cirrosis Hepática')
ax.bar('Embarazo', num_embarazo, 0.5, label='Embarazo')

ax.set_xlabel('Número de Casos')
ax.set_ylabel('Condición Médica')
ax.set_title('Número de Casos de Dengue por Condición Médica')
# Agregar las cantidades encima de cada barra
for i, v in enumerate([num_hemorragicos, num_diabetes, num_hipertension, num_ulc_peptica, num_enfermedad_renal,num_inmunosupr, num_cirrosis_hepatica, num_embarazo]):
    ax.annotate(str(v), xy=(i, v), xytext=(0, 3), textcoords="offset points", ha='center')
# a 45 grados
plt.xticks(rotation=45, ha='right')
# Colocar la leyenda en la esquina superior derecha
ax.legend(loc='upper right')
plt.show()
```

```python
# calcula el porcentaje de personas fallecidas y no fallecidas utilizando el campo "DEFUNCION"
labels = ['No Fallecidos','Fallecidos' ]
colors = ['grey', 'red']
# Imprimir cantidad de fallecidos por dengue hemorrágico
fallecidos = df[df['DEFUNCION'] == 1]['HEMORRAGICOS'].value_counts()
print("Fallecidos:{} ".format(fallecidos))

# Crear la figura con un tamaño específico
plt.figure(figsize=(3, 3))
# Resaltar la porción de fallecidos
explode = (0.1, 0)  
# Graficar pastel
plt.pie(fallecidos, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Distribución de Muertes por Dengue Hemorrágico')
plt.axis('equal')
plt.show()
```

```python
# Get a percentage of hemorrhagic and classic dengue cases and plot it in a pie chart
# Calculate the total number of cases
total_cases = df['HEMORRAGICOS'].count() # Get all the values of the column HEMORRAGICOS

# Calculate the number of hemorrhagic dengue cases
hemorrhagic_cases = df[df['HEMORRAGICOS'] == 2]['HEMORRAGICOS'].count() # Use 2 times the column HEMORRAGICOS, firts to get the number of cases with number 2 and then to get the number of values

# Calculate the number of classic dengue cases
classic_cases = df[df['HEMORRAGICOS'] == 1]['HEMORRAGICOS'].count() 

# Function to calculate the percentage of a type of dengue
def get_percentage(type_dengue, all_cases):
    percentage = (type_dengue / all_cases) * 100
    return percentage

hemorrhagic_percentage = get_percentage(hemorrhagic_cases, total_cases)
classic_percentage = get_percentage(classic_cases, total_cases)

# Create a pie chart
labels = ['Hemorrhagic Dengue', 'Classic Dengue']
sizes = [hemorrhagic_percentage, classic_percentage]
colors = ['red', 'green']
explode = (0.1, 0)  # explode the first slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Percentage of Hemorrhagic Dengue and Classic Dengue Cases')

plt.show()
```

```python
# determinar el porcentaje de pacientes de las diferentes enfermedades en el dataset

# Calcular el porcentaje de pacientes con diabetes
porcentaje_diabetes = (df['DIABETES'].sum() / df['DIABETES'].count()) * 100
porcentaje_hipertension = (df['HIPERTENSION'].sum() / df['HIPERTENSION'].count()) * 100
porcentaje_enfermedad_ulc_peptica = (df['ENFERMEDAD_ULC_PEPTICA'].sum() / df['ENFERMEDAD_ULC_PEPTICA'].count()) * 100
porcentaje_renal = (df['ENFERMEDAD_RENAL'].sum() / df['ENFERMEDAD_RENAL'].count()) * 100
porcentaje_inmunosupr = (df['INMUNOSUPR'].sum() / df['INMUNOSUPR'].count()) * 100
porcentaje_cirrosis = (df['CIRROSIS_HEPATICA'].sum() / df['CIRROSIS_HEPATICA'].count()) * 100
porcentaje_embarazo = (df['EMBARAZO'].sum() / df['EMBARAZO'].count()) * 100

# Crear un gráfico de pastel
# Crear una lista con los porcentajes de las enfermedades adicionales
porcentajes_enfermedades = [porcentaje_diabetes, porcentaje_hipertension, porcentaje_enfermedad_ulc_peptica, porcentaje_renal, porcentaje_inmunosupr, porcentaje_cirrosis, porcentaje_embarazo]

# Crear una lista con los nombres de las enfermedades adicionales
enfermedades = ['Diabetes', 'Hipertensión', 'Enfermedad Ulcera Péptica', 'Enfermedad Renal', 'Inmunosupresión', 'Cirrosis Hepática', 'Embarazo']

# Crear la gráfica de pastel
plt.figure(figsize=(6, 6))
plt.pie(porcentajes_enfermedades, labels=enfermedades, autopct='%1.1f%%', startangle=90)

# Configurar la gráfica
plt.title('Porcentaje de Enfermedades Adicionales')
plt.axis('equal')

# Mostrar la gráfica
plt.show()

```
